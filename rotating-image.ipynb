```python
import cv2 
import numpy as np 
  
ROTATING_IMAGE_SHAPE = (800, 1200)
def detecting_circles(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    gray_blurred = cv2.blur(gray, (3, 3)) 
     

    # Apply Hough transform on the blurred image. 
    detected_circles = cv2.HoughCircles(gray_blurred,  
                    cv2.HOUGH_GRADIENT, 1, 20, param1 = 50, 
                param2 = 30, minRadius = 1, maxRadius = 20) 

    coordinates = []
    if detected_circles is not None: 
        detected_circles = np.uint16(np.around(detected_circles)) 

        for pt in detected_circles[0, :]: 
            a, b, r = pt[0], pt[1], pt[2] 
            coordinates.append((a, b))
            
    return coordinates
```

```python
import os

directory = "./"
os.chdir(directory)
path = "dataset/"

for filename in os.listdir(os.path.join(path, "Train")):
    full_file_name = os.path.join(path, "Train", filename)
    
    # Reading image
    img = cv2.imread(full_file_name, cv2.IMREAD_COLOR) 
    img = cv2.resize(img, ROTATING_IMAGE_SHAPE)
    
    circle_coordinates = detecting_circles(img)
    # if len(circle_coordinates) < 1:
    #     continue
    left_count = sum(1 for coord in circle_coordinates if coord[0] < 400)
    if left_count > len(circle_coordinates) / 2:
        img = cv2.rotate(img, cv2.ROTATE_180)
    
    print(f"{filename}\nCoordinates: {circle_coordinates}")
    cv2.imwrite(os.path.join("./dataset-rotated", "Train", filename), img)
```

```python
import cv2 
import numpy as np 
img = cv2.imread("dataset/Train/TPS_392.jpg", cv2.IMREAD_COLOR) 

img = cv2.resize(img, (800, 1200))
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 
gray_blurred = cv2.blur(gray, (3, 3)) 

# Apply Hough transform on the blurred image. 
detected_circles = cv2.HoughCircles(gray_blurred,  
                cv2.HOUGH_GRADIENT, 1, 20, param1 = 50, 
            param2 = 30, minRadius = 1, maxRadius = 20) 

coordinates = []
if detected_circles is not None: 
    detected_circles = np.uint16(np.around(detected_circles)) 

    for pt in detected_circles[0, :]: 
        a, b, r = pt[0], pt[1], pt[2] 
        coordinates.append((a, b))
        
        # Draw the circumference of the circle. 
        cv2.circle(img, (a, b), r, (0, 255, 0), 2) 
  
        # Draw a small circle (of radius 1) to show the center. 
        cv2.circle(img, (a, b), 1, (0, 0, 255), 3) 
cv2.imshow("Detected Circle", img) 
cv2.waitKey(0)
print(coordinates)
```
